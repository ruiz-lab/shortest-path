```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def showFig(dir):
    graph_sizes = []
    n_nodes = 10
    max_n_nodes = 1280*32
    while n_nodes <= max_n_nodes:
        graph_sizes.append(n_nodes)
        n_nodes = n_nodes*2
    if 'lbd' in dir:
        num_plots = len(graph_sizes)*6
        train_index = (range(num_plots)[2::6])
        val_index = (range(num_plots)[3::6])
        test_index = (range(num_plots)[4::6])
        dist_index = (range(num_plots)[5::6])
        for i in range(len(graph_sizes)):
            print(graph_sizes[i])
            fig = plt.figure(figsize=(20, 10))
            img = mpimg.imread('outputs/'+dir+'/'+str(train_index[i])+'.png')
            fig.add_subplot(1, 4, 1)
            plt.imshow(img)
            plt.axis('off')
            img = mpimg.imread('outputs/'+dir+'/'+str(val_index[i])+'.png')
            fig.add_subplot(1, 4, 2)
            plt.imshow(img)
            plt.axis('off')
            img = mpimg.imread('outputs/'+dir+'/'+str(test_index[i])+'.png')
            fig.add_subplot(1, 4, 3)
            plt.imshow(img)
            plt.axis('off')
            img = mpimg.imread('outputs/'+dir+'/'+str(dist_index[i])+'.png')
            fig.add_subplot(1, 4, 4)
            plt.imshow(img)
            plt.axis('off')
            plt.tight_layout()
            plt.show()
    else:
        if dir == 'dump3':
            num_plots = len(graph_sizes)*4
            train_index = (range(num_plots)[1::4])
            val_index = (range(num_plots)[2::4])
            test_index = (range(num_plots)[3::4])
        else:
            num_plots = len(graph_sizes)*5
            train_index = (range(num_plots)[2::5])
            val_index = (range(num_plots)[3::5])
            test_index = (range(num_plots)[4::5])
        for i in range(len(graph_sizes)):
            print(graph_sizes[i])
            fig = plt.figure(figsize=(15, 30))
            img = mpimg.imread('outputs/'+dir+'/'+str(train_index[i])+'.png')
            fig.add_subplot(1, 3, 1)
            plt.imshow(img)
            plt.axis('off')
            img = mpimg.imread('outputs/'+dir+'/'+str(val_index[i])+'.png')
            fig.add_subplot(1, 3, 2)
            plt.imshow(img)
            plt.axis('off')
            img = mpimg.imread('outputs/'+dir+'/'+str(test_index[i])+'.png')
            fig.add_subplot(1, 3, 3)
            plt.imshow(img)
            plt.axis('off')
            plt.tight_layout()
            plt.show()
```

```python
showFig('lbd 1')
```
